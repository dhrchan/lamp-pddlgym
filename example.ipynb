```python
import pddlgym

env = pddlgym.make("PDDLEnvProb_blocks-v0")
env.fix_problem_index(1)
```

```python
from lamp import GoalNetwork

goal_network = GoalNetwork(env)
goal_network.show_goal_network()
```

```python
from lamp import LAMPPlanner, LAMPConfig

cfg = LAMPConfig(
    n_rollouts = 500, 
    horizon = 20, 
    budget = 100,
    exploration_const = 2 ** 0.5, 
    normalize_exploration_const = False, 
    n_init = 5,
    greediness = 0.5,
    risk_factor = -0.1, 
    goal_utility = 1, 
    h_util = lambda _: 1, 
    h_ptg = lambda _: 1,
    seed = 1, 
    show_progress = False
)

planner = LAMPPlanner(cfg)

planner.run(env, show_progress=True)
```
